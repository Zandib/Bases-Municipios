## Nome, grau de escolaridade, genero

```python
df_candidatos.columns = [unidecode(c).lower().strip() for c in df_candidatos.columns]
```

```python
df_candidatos = df_candidatos.loc[df_candidatos['ds_cargo']=='PREFEITO']
```

```python
df_candidatos.columns
```

```python
colunas_caracteristicas  = ['ds_genero','ds_grau_instrucao','ds_estado_civil','ds_cor_raca','ds_ocupacao']
colunas_chave = ['nr_candidato','nm_candidato','nm_ue','sg_ue']
```

```python
resumo_candidato = df_candidatos.groupby(colunas_chave)[colunas_caracteristicas].first().reset_index()
```

```python
resumo_candidato.head()
```

## Eleito e não eleito e quantidade de votos

```python
df_votacao_zona.columns = [unidecode(c).lower().strip() for c in df_votacao_zona.columns]
```

```python
df_votacao_zona = df_votacao_zona.loc[df_votacao_zona['ds_cargo']=='Prefeito']
```

```python
df_votacao_zona['ds_sit_tot_turno'].unique()
```

```python
df_votacao_zona.head()
```

```python
colunas_caract = ['ds_sit_tot_turno','qt_votos_nominais_validos']
colunas_chave = ['nr_candidato','nm_candidato','nm_ue','sg_ue','nr_turno']
```

```python
resumo_votacao = df_votacao_zona.groupby(colunas_chave).agg({
    'ds_sit_tot_turno': 'first',
    'qt_votos_nominais_validos': 'sum'
}).reset_index()
```

```python
resumo_votacao.loc[resumo_votacao['nm_ue']=='SÃO PAULO'].sort_values('qt_votos_nominais_validos')
```

```python
resumo_votacao = resumo_votacao.sort_values('nr_turno',ascending=False)
```

```python
turno_mapper = resumo_votacao.groupby(['nm_ue','sg_ue'])['nr_turno'].max().reset_index()
```

```python
resumo_votacao = resumo_votacao.merge(turno_mapper,on=['nm_ue','sg_ue','nr_turno'],how='inner')
```

```python
resumo_votacao
```

## Resumo do municipio

```python
resumo_municipio = df_votacao_zona.groupby(['nm_ue','sg_ue'])['qt_votos_nominais_validos'].sum().reset_index()
```

```python
resumo_municipio.rename(columns={"qt_votos_nominais_validos":'total_votos_municipio'},inplace=True)
```

```python
resumo_municipio
```

## Join das infos

```python
resumo_completo = resumo_votacao.merge(
    resumo_candidato
    , on = ['nr_candidato','nm_candidato','nm_ue','sg_ue']
    , how = 'left'
)
```

```python
resumo_completo = resumo_completo.merge(
    resumo_municipio
    , on = ['nm_ue','sg_ue']
    , how = 'inner'
)
```

### Limitar a dois candidatos

```python
resumo_completo = resumo_completo.sort_values('qt_votos_nominais_validos',ascending=False)
```

```python
resumo_completo = resumo_completo.groupby(['nm_ue','sg_ue']).head(2)
```

```python
resumo_completo.sort_values(['qt_votos_nominais_validos'],ascending=False).head(10)
```

```python
resumo_completo['%'] = resumo_completo['qt_votos_nominais_validos']/resumo_completo['total_votos_municipio']
```

```python
resumo_completo.head()
```

## Pivot table

```python
df_eleito = resumo_completo.loc[resumo_completo['ds_sit_tot_turno']=='ELEITO']
df_n_eleito = resumo_completo.loc[resumo_completo['ds_sit_tot_turno']=='NÃO ELEITO']
```

```python
resumo_pivot = df_eleito.merge(
    df_n_eleito
    , on = ['nm_ue','sg_ue']
    , how = 'left'
    , suffixes = ('_eleito','_nao_eleito')
)
```

```python
resumo_pivot
```

```python
resumo_pivot['diferenca_pp'] = (resumo_pivot['%_eleito']-resumo_pivot['%_nao_eleito'])
```

```python
resumo_pivot['flag_dif_5pp'] = (resumo_pivot['diferenca_pp']<=.05).astype(int)
resumo_pivot['flag_dif_10pp'] = (resumo_pivot['diferenca_pp']<=.1).astype(int)
```

```python
resumo_pivot['flag_genero_diferente_eleito_nao_eleito'] = \
(resumo_pivot['ds_genero_eleito'] != resumo_pivot['ds_genero_nao_eleito']).astype(int)
```

```python
resumo_pivot = resumo_pivot.merge(depara_municipio,on=['sg_ue'],how='left',suffixes=('','_depara'))
```
